```python
from igraph import *
import cairocffi
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
```

```python
file_path = "./data/karate.adjlist"

def load_adjlist(filename, directed=True):
    edgelist = []
    names = UniqueIdGenerator()
    for line in open(filename):
        parts = line.strip().split()
        u = names[parts.pop(0)]
        edgelist.extend([(u, names[v]) for v in parts])
    g = Graph(edgelist, directed=directed)
    g.vs["name"] = names.values()
    return g

nx_graph = nx.readwrite.adjlist.read_adjlist(file_path, create_using=nx.DiGraph())
ig_graph = load_adjlist(file_path)
```

# Introduction
In this example notebook, we want to implement DeepWalk algorithm for graph embedding. Zachry's karate club dataset. It is a directed graph and plotted as following.

```python
plt.figure(figsize=(10, 10))
nx.draw(nx_graph, cmap = plt.get_cmap('jet'), with_labels = True, node_color='red')
```

# Generate path and training data

```python
from keras.preprocessing.sequence import skipgrams
```

```python
def batch_generator(file_path = file_path, window_size = 4, path_len = 10):
    nx_graph = nx.readwrite.adjlist.read_adjlist(file_path, create_using=nx.DiGraph())
    ig_graph = load_adjlist(file_path)
    vocab_size = len(nx_graph.nodes)
    sentences = [ig_graph.random_walk(k, path_len) for k in range(vocab_size)]
    for sent in sentences:
        couples, labels = skipgrams(sent, vocab_size, window_size = window_size, negative_samples = 1.0)
        yield couples, labels
```

# Build model
In this section, we borrow the model we used in word2vec example and apply it directly to this case. We will use __negative sampling__ and binary loss.

```python
from keras.models import Model
from keras.layers import Input, Flatten, Embedding, dot, Activation
```

```python
def build_model(input_dim, embedding_dim):
    target_input_token = Input(shape=[1])
    context_input_token = Input(shape=[1])
    
    target_embed = Embedding(output_dim = embedding_dim, \
                             input_dim = input_dim, input_length = 1, \
                             name = "target_embed")(target_input_token)
    context_embed = Embedding(output_dim = embedding_dim, \
                             input_dim = input_dim, input_length = 1, \
                             name = "context_embed")(context_input_token)
    
    target_vec = Flatten()(target_embed)
    context_vec = Flatten()(context_embed)
    
    y = Activation('sigmoid')(dot([target_vec, context_vec], axes = 1))
    model = Model(inputs=[target_input_token, context_input_token], outputs = [y])
    model.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop')
    
    return model
```

```python
def train_model(model, file_path = file_path, epochs = 3):
    for epoch in range(epochs):
        loss = 0
        print("{}/{} epochs".format(epoch, epochs))
        for couples, labels in batch_generator(file_path = file_path):
            words, contexts = zip(*couples)
            words = np.array(words, dtype=np.int32)
            contexts = np.array(contexts, dtype=np.int32)
            y = np.array(labels, dtype=np.int32)
            loss += model.train_on_batch([words, contexts], y)
        print("loss: {}".format(loss))
    print("Training finished")
```

# Train model

```python
node_num = len(nx_graph.nodes)
model = build_model(node_num, 2)
model.summary()
```

```python
train_model(model, file_path)
```

```python
vectors = model.get_weights()[0]
```
